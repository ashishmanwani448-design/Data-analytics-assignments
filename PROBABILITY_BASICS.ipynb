## Question 1

A die is rolled. What is the probability of getting:

### (a) An even number

ANS :

**Sample space:**  
$$
S = \{1,2,3,4,5,6\}
$$

**Even numbers:**  
$$
\{2,4,6\}
$$

**Probability:**
$$
P(\text{Even}) = \frac{3}{6} = \frac{1}{2}
$$

$$
\{5,6\}
$$

**Probability:**
$$
P(>4) = \frac{2}{6} = \frac{1}{3}
$$

## Question 2 In a class of 50 students:

- 20 like Mathematics (M)  
- 15 like Science (S)  
- 5 like both subjects  

What is the probability that a student chosen at random likes Mathematics or Science?

ANS:



**Probability that a student likes Mathematics or Science:**

$$
P(M \cup S) = \frac{20 + 15 - 5}{50}
$$

$$
= \frac{30}{50} = \frac{3}{5}
$$

## Question 3  A bag has 3 red and 2 blue balls. If one ball is drawn randomly and is red, what is the probability that the next ball is also red (without replacement)?

ANS:



A bag has 3 red and 2 blue balls.

One red ball is drawn first **without replacement**.

**Remaining balls:**  
- Red = 2  
- Total = 4  

**Probability that the next ball is also red:**

$$

## Question 5  The average height of 1000 students = 160 cm. A sample of 100 students shows an average height = 158 cm. Find the sampling error.


ANS :


- Population mean height = 160 cm  
- Sample mean height = 158 cm  

**Sampling error:**

$$
\text{Sampling Error} = \text{Sample Mean} - \text{Population Mean}
$$

$$
= 158 - 160 = -2 \text{ cm}
$$

## Question 6 The population mean salary is ₹50,000 with σ = ₹5,000. If we take a sample of 100 employees, what is the standard error of the mean (SEM)?


ANS :


- Population mean salary = ₹50,000  
- Standard deviation (σ) = ₹5,000  
- Sample size (n) = 100  

**Standard Error of Mean (SEM):**

$$
SEM = \frac{\sigma}{\sqrt{n}}
$$

$$
= \frac{5000}{\sqrt{100}} = \frac{5000}{10} = 500
$$

**Answer:**  
SEM = **₹500**

## Question 7  In a group of 100 students:

- 40 like Cricket (C)
- 30 like Football (F)
- 10 like both Cricket and Football

Find the probability that a student likes at least one sport.

ANS :



In a group of 100 students:

- 40 like Cricket (C)  
- 30 like Football (F)  
- 10 like both  

**Probability that a student likes at least one sport:**

$$
P(C \cup F) = \frac{40 + 30 - 10}{100}
$$

$$
= \frac{60}{100} = 0.6
$$

## Question 8 From a deck of 52 cards, two cards are drawn without replacement. What is the probability that both are Aces?

ANS :    



From a deck of 52 cards, two cards are drawn without replacement.

**Probability that both are Aces:**

$$
P = \frac{4}{52} \times \frac{3}{51}
$$

$$
= \frac{12}{2652} = \frac{1}{221}

## Question 10 Differentiate between discrete and continuous random variables with examples.

ANS :    



### Difference between Discrete and Continuous Random Variables

**Discrete Random Variable:**  
- Takes countable values  
- Example: Number of students in a class, number of heads in coin tosses  

**Continuous Random Variable:**  
- Takes infinite values within a range
