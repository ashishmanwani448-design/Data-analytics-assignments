## Easy Level

### Q1: Q1: Understanding Central Tendency (Easy)
A bakery tracks the daily sales of muffins (in dozens) over a week: [10, 12, 11, 15, 14, 13, 12].
What is the most representative value of their weekly sales, and why?


ANS:

**Given data (daily sales in dozens):**  
10, 12, 11, 15, 14, 13, 12  

**Mean calculation:**

$$
\text{Mean} = \frac{10+12+11+15+14+13+12}{7}
$$

$$
= \frac{87}{7} = 12.43
$$

**Answer:**  
The most representative value is the **mean = 12.43 dozens**, because it includes all observations.

### Q9: Combining Measures (Hard)
A family records their monthly electricity usage (in kWh): [400, 420, 390, 450, 410]. Find the mean and standard deviation. What do these values together tell you about the family’s energy use
pattern?



ANS:



**Electricity usage (kWh):**  
400, 420, 390, 450, 410  

**Mean calculation:**

$$
\text{Mean} = \frac{400+420+390+450+410}{5}
$$

$$
= \frac{2070}{5} = 414
$$

**Standard Deviation (Population):**

$$
\text{Variance} = \frac{2120}{5} = 424
$$

$$
\text{Standard Deviation} = \sqrt{424} = 20.59
$$

**Answer:**  
Mean = **414 kWh**, SD = **20.59**, showing stable energy usage.

### Q10: Practical Application (Hard)
A basketball player’s points in 8 games are recorded: [15, 18, 20, 22, 25, 17, 19, 21].
Find the mean, median, mode, range, and standard deviation. What insights can these measures provide
about the player’s scoring performance?


ANS:



**Points scored:**  
15, 18, 20, 22, 25, 17, 19, 21  

**Mean:**

$$
\text{Mean} = \frac{157}{8} = 19.63
$$

**Median:**

$$
\text{Median} = \frac{19+20}{2} = 19.5
$$

**Mode:**  
No repeated value → **No mode**

**Range:**

$$
\text{Range} = 25 - 15 = 10
$$

**Standard Deviation (Population):**

$$
\text{SD} \approx 3.16
$$

**Answer:**  
Mean = **19.63**, Median = **19.5**, Mode = **None**,  
Range = **10**, SD = **3.16**  

These values indicate consistent scoring performance.
